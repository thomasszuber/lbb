```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tight = pd.read_csv("../bases/rome_tension.csv")
```

```python
tight = tight.groupby('rome5').T.mean()
tight = pd.DataFrame({'rome':tight.index,'T':tight.values})
```

```python
herf = pd.read_csv("../bases/naf_rome.csv")
```

```python
herf = herf.groupby('rome').herf_rome.mean()
herf = pd.DataFrame({'rome':herf.index,'H':herf.values})
```

```python
tight_herf = pd.merge(tight,herf,how='left',on='rome')
```

```python
tight_herf['T'].corr(tight_herf['H'])
```

```python
plt.scatter(tight_herf['H'],tight_herf['T'])
plt.savefig('tension_herf.png')
```

```python
tight_herf['cat'] = tight_herf['rome'].apply(lambda x: x[0])
tight_herf['div'] = tight_herf['rome'].apply(lambda x: x[0:2])
```

```python
plt.figure(figsize=[15,15])
plt.subplots()
for i,k in enumerate('ABCDFGHIJKLMN'):
    y = tight_herf.loc[tight_herf.cat == k]
    plt.subplot(5,3,i+1)
    plt.scatter(y['H'],y['T'])
    plt.title(k)
```
